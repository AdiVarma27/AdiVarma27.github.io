## Intuition behind Hypothesis Testing 

* This post covers the concepts of Hypothesis Testing (Z-test, t-test, proportions: one-tailed, two-tailed tests).
* It takes a deeper look at Type-1, Type-2 errors, along with the infamous <strong><em>p-value</em></strong> (No, it is not the probability of Type-1 error).
* Before we start, lets take a look at the Normal Distribution, followed by the Standard Normal Distribution.

### Normal Distribution 

* The normal distribution is probably the most commonly occuring continuous probability distribution we come across. Unlike uniform distribution, where each entity has equally distributed occurance/ probability, most of the occurances are grouped at an average value, with lesser occurances on either sides of the average value. 
* We see this phenomenon in everyday life; patterns which follow the Central Limit Theorem. The number of hours you sleep every night, the number of calories you eat in a day, the number of steps you take everyday, which are everyday things we do.
* The PDF of the Normal Distribution (with mean ùúá, variance ùúé2) is as follows:

$$ {\displaystyle f(x\mid \mu ,\sigma ^{2})={\frac {1}{\sqrt {2\pi \sigma ^{2}}}}e^{-{\frac {(x-\mu )^{2}}{2\sigma ^{2}}}}}$$

```python
# Using scipy norm function

import math
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt


def NormalDistPlot(mean, variance, lower_limit, upper_limit, color):
    axis_range = np.arange(lower_limit, upper_limit, 0.01)
    plt.plot(axis_range, norm.pdf(axis_range, mean, variance),color)
    plt.show()
    
    
NormalDistPlot(10, 1, 5, 15, color='blue')
NormalDistPlot(10, 6, 5, 15, color='red')

# The plot below in blue shows a normal distribution with mean 10 and variance 2.
# The plot below in red shows a normal distribution with mean 10 and variance 6.
```

#### Standard Normal Distribution
* The standard normal is a simple case of a normal distribution, with mean at 0, and variance 1. Below is the plot of Standard Normal. We use the standard normal to explain the concept of Z-score easily. 

#### Z-score or Standard Score
* The z-score or the standard score, explain many deviations away a point is from the mean. Note that this value is dimensionless, and can be thought of as a <strong><em>'normalization'</em></strong> step for the data-points.
* Z-score of any point x, can be calculated by using the following formula: $$z={(x-\mu)  \over \sigma }$$<br>
* The Standard Normal distribution follows the <strong><em>68-95-99 rule</em></strong>, also known as the <strong><em>emperical rule.</em></strong> According to this rule, 68 % of data points lie within 1-standard deviation away from the mean, 95 % of data points lie within 2 standard deviations away from the mean, and 99.7 % of data points lie within 3 starndard deviations away from the mean.

* What does Z-score tell us ? Apart from the formal definition, it also gives us the confidence; a point with z-score of 0, means the point lies exactly on the mean, and has the highest probability of occuring. Hence, it also denotes confidence of a point belonging to the distribution.

```python
plt.figure(figsize=(12,8))

plt.plot(np.arange(-3, -2, 0.01), norm.pdf(np.arange(-3, -2, 0.01), 0, 1),'red')
plt.plot(np.arange(-2, -1, 0.01), norm.pdf(np.arange(-2, -1, 0.01), 0, 1),'orange')
plt.plot(np.arange(-1, 0, 0.01), norm.pdf(np.arange(-1, 0, 0.01), 0, 1),'green')

plt.plot(np.arange(0, 1, 0.01), norm.pdf(np.arange(0, 1, 0.01), 0, 1),'green')
plt.plot(np.arange(1, 2, 0.01), norm.pdf(np.arange(1, 2, 0.01), 0, 1),'orange')
plt.plot(np.arange(2, 3, 0.01), norm.pdf(np.arange(2, 3, 0.01), 0, 1),'red')

plt.fill_between(np.arange(-3,-2,0.01), norm.pdf(np.arange(-3, -2, 0.01), 0, 1), 0, 
                 color='red',       
                 alpha=0.5) 

plt.fill_between(np.arange(2,3,0.01), norm.pdf(np.arange(2, 3, 0.01), 0, 1), 0, 
                 color='red',       
                 alpha=0.5) 

plt.fill_between(np.arange(-2,-1,0.01), norm.pdf(np.arange(-2, -1, 0.01), 0, 1), 0, 
                 color='orange',       
                 alpha=0.4) 

plt.fill_between(np.arange(1,2,0.01), norm.pdf(np.arange(1, 2, 0.01), 0, 1), 0, 
                 color='orange',       
                 alpha=0.4) 

plt.fill_between(np.arange(-1,1,0.01), norm.pdf(np.arange(-1, 1, 0.01), 0, 1), 0, 
                 color='green',       
                 alpha=0.2)

plt.fill_between(np.arange(-0.01,0.01,0.01), norm.pdf(np.arange(-0.01, 0.01, 0.01), 0, 1), 0, 
                 color='green',       
                 alpha=0.3)

plt.xlabel('Standard Deviation')
plt.ylabel('Probability')

plt.xlim(-3,3)
plt.ylim(0,0.5)

plt.show()
```

* From the above visual, we see that 68 % data lies within the green region, 95 % lies within both exremes of orange region (including green), and 99.7 % of data lies within red region (all colors).
* We see that the colors shows the amount of <strong><em>'risk'</em></strong> of some sort, associated with data points belonging to each group. How do we deem them to be risky ? In what sense are they worth taking a deeper look ?

### Hypothesis Testing
* Any claim we make, needs to be backed up by numbers. Let's stick to a simple yet relatable example. My personal favourite sandwich at Chick-Fil-A, is the chick-fil-a deluxe sandwich (with cheese). According to their website, this sandwich with cheese, is worth 500 Calories, 23 grams of fat, 42 grams of carbs and 31 grams of protein. (https://www.chick-fil-a.com/menu-items/chick-fil-a-deluxe-sandwich).

* However, I believe that there is no way that they can pack in exactly 500 Calories per sandwich. Sometimes I feel the sandwich is much larger, sometimes the thickness of the patty feels much leaner. (It's how I feel, without any valid proof or backing up my theory with numbers). Let's say I want to prove my claim, that the sandwich cannot be 500 calories. Let's assume a standard deviation of 15 calories.

### My Experiment Setup:

* Any claim or hypothesis that we would want to test, is generally speaking, the <strong><em>Alternate Hypothesis</em></strong>. Hence, the <strong><em>Null Hypothesis</em></strong>, is that the sandwich is 500 calories. 

* To prove my points, Let's say I set up an experiment. Every single time I head out to Chick-Fil-A, I buy an extra sandwich, which is put into a calorimeter, which tells me the number of calories accurately and precisely. Over 6-months, I went to Chick-Fil-A 64 times, and got 64 extra sandwiches. I make note of all the results from my experiment results, and find out that the mean calorie count is 508 calories.

* What we know till now: The company claims to serve sandwiches with 500 calories. 
$${{H0: ùúá = 500}}$$ $${{Ha: ùúá != 500}}$$ 

* The <strong><em>Null Hypothesis</em></strong> states that the sandwich is 500 calories, and the <strong><em>Alternate Hypothesis</em></strong> (my claim which I'm testing), is that they do not serve 500 calories.

* We know the experimental mean calories, 508 calories. Now, my sample size of 64 sandwiches, is just a small sample of the actual total population. However, according to Law of Large Number (As n->infinity, population mean and sample mean converge). The standard error of the mean (https://en.wikipedia.org/wiki/Standard_error) or the Standard Deviation of the sampling disribution is as follows: $${{\sigma} (sample)} = {\frac {\sigma (population)}{\sqrt {n}}}$$

* How did we arrive to the conclusion that the Standard error term is Population deviation divided by square root of n ? Remember, our 64 trials are actually a sampling distribution of the actual population sample mean. It might seem a bit confusing initially, here's some clarification. (SOURCE: https://stats.stackexchange.com/questions/89154/general-method-for-deriving-the-standard-error)

* The Chick-Fil-A branch near me makes around a thousand sandwiches every day, and this branch opened up around 2 years ago. That's over 10 Million sandwiches over 3 years. If we actaully counted those 10 million and plotted a PDF of calories (remember, we have no idea how this looks), but we assume it to be normally distributed. Their claim of 500 calories for each sandwich comes from the distribution we just described, which is the total population. 

* My sample (64 burgers), is a random sample with n=64, and the experimental mean truns out to be 508. We know from basic intuition, that smaller the sample size, smaller the sample size, higher is the deviation. 

### Confidence Intervals

* We are ready to accept some level of error, hence, we have a confidence interval; generally, most problems tend to choose 95 % Confidence interval. The confidence intervals are the points within which 95 % of the data exists. From the above diagram, we remember that 95 % of data lies within 2 standard deviations (1.96 Standard deviations to be exact).

* Hence, for the sampling distribution (with mean 500 calories), the Confidence interval is as follows: $${{[x - 1.96 {\sigma(population) \over {\sqrt n}}, x + 1.96 {\sigma(population) \over {\sqrt n}}]}}$$ 
 $$[500 - 1.96*{15 \over {\sqrt 64}}, 500 + 1.96*{15 \over {\sqrt 64}}] =[496.325, 503.675] $$

```python
plt.figure(figsize=(12,8))

plt.plot(np.arange(485, 515, 0.01), norm.pdf(np.arange(485, 515, 0.01), 500, (15/math.sqrt(64))),'red')

plt.fill_between(np.arange(485, 496.325, 0.01), norm.pdf(np.arange(485, 496.325, 0.01), 500, (15/math.sqrt(64))), 0, 
                 color='red',       
                 alpha=0.3)

plt.fill_between(np.arange(503.675, 515, 0.01), norm.pdf(np.arange(503.675, 515, 0.01), 500, (15/math.sqrt(64))), 0, 
                 color='red',       
                 alpha=0.3)

plt.fill_between(np.arange(499.9, 500.1, 0.01), norm.pdf(np.arange(499.9, 500.1, 0.01), 500, (15/math.sqrt(64))), 0, 
                 color='green',       
                 alpha=0.3)

plt.fill_between(np.arange(507.9, 508.1, 0.01), norm.pdf(np.arange(507.9, 508.1, 0.01), 500, (15/math.sqrt(64))), 0.2, 
                 color='green',       
                 alpha=1)
```

### Was my experiment a success ?

* The region in red on both sides, shows the Confidence Interval, beyond which only 5 % of the data exists. (Points below z-score of -1.96 and points above z-score of 1.96). Onto the right, we see the experimental mean at 508 calories, and null-hypothesis mean at 500 calories. 
* We can directly <strong> <em> reject the null hypothesis </em> </strong> (That the sandwiches have 500 calories on average), as clearly the experimental mean (508 calories), is at least 1.96 Standard deviations away from the mean. (Remember, 1.96 Standard deviations has 95 % of the data of the normal distribution).

### How many standard deviations away is the sample mean  of 508 calories ? 

* This is where the concept of Z-score plays a role, to see how far the experimental mean is away from the Null Hypothesis mean. (How many standard deviations 508 is from 500 calories).

* Hence, we are normalizing the number to unit standard deviation scale by applying the transformation. $$z={x-\mu  \over \sigma } = {508 - 500 \over {15 \over {\sqrt {64}}}}  = 4.266$$ 

* The Z-score of the mean (508 calories) is 4.266, which is 4.266 Standard deviations away from the actual population distribution. This is way greater than the 1.96 Standard deviations. Hence, my claim holds true !

* We can conclude that by finding Confidence Intervals, as well as using Z-stat, we can decide to rejeect the null or not. (If the company claim is correct, or mine is), then what is <strong> <em> p-value </em> </strong>? 

* Before we talk about p-value, let us talk about left-tail and right-tail tests. While forming Hypothesis, context is everything. In our previous example, we conducted a two-tailed test; Why was it a 2-tailed test ? 
* My claim was that the number of calories are not 500. It meant that number of calories could be significantly lesser, or significantly higher; hence, there could be two rejection regions as well see throughout all the diagrams above.

### Left Tail Test
* For example, let us consider water bottle company claims to package 350 ml of water in each bottle. A few consumers observe that the water packaging is not consistent, and sometimes find less than 350 ml of water, ehcen want to test their claim. 

* In this case, the null hypothesis is that average volume of water in a bottle is 350 ml, alternate hypothesis is that it is less than 350 ml. Hence, this falls under the category of a left tail test.

### Right Tail Test
* For example, let us consider a new study which shows that taking a new medicine helps patients sleep better. We want to measure the effectiveness of this medicine. One group sleeps for 7 hours, and a different group sleep (which takes the medicine), sleeps for 7.5 hours. Hence, we are testing the hypothesis that the new medicine helps sleep better, and the cutoff region is towards right hand side, hence is a right tailed test.

### The fuss about p-value
* As you see below, and with the z-stat of 4.266, we know that the alternate hypothesis is true (we reject the null hypothesis). Another way to come to this conclusion is by using the p-value.
* By plugging in the Z-stat into p-value table (in any stats textbook or online chart), we get pvalue of 0.00002. As this is a two-tailed test, each side of he tail has 0.00001 of the area under distribution. 
* As <strong> p-value (0.00002) <  Alpha (0.05) </strong>, we can reject the null hypothesis. <strong> Essentially, p-value is the probability that finding the observed or even more extreme results, under the assumption that Null Hypothesis is True.</strong>
* The significance level (0.05- alpha) and p-value are <strong>NOT</strong> the same. The significance value is something that we choose (generally 0.05 or sometimes even 0.01), and compare the p-value with 0.05. If the p-value is greater than 0.05, that means there is a significant chance that the null hypothesis is true (due to sample meanbeing close to the null hypothesis mean). Hence, we could use Confidence Intervals, Z-stat or p-value to decide our Hypothesis Tests outcome.

### Type-I and Type-II errors 
* Type-I error occurs when the null hypothesis is incorrectly rejected, Type-II error occurs when we inccorectly accept the null hypothesis.
* To make things simple (I can never forget the difference between Type-I and Type-II erros after seeing this picture). <strong> A picture is worth a thousand words.</strong> SOURCE: https://stats.stackexchange.com/questions/110433/examples-for-type-i-and-type-ii-errors

![alt text](s2p1.png "Difference between Type-I and Type=II error")

* For each hypothesis test, we always start by choosing the acceptable rate of Type-I error (0.05 in our case). The rate of Type-II error can be controlled by varying the rate of Type-I error. 
* In different context, type-I and type-II errors can have different costs. For example, in treating deadly diseases, Type-II error needs to be minimized (Chance of not able to detect if an individual has cancer) should ideally be zero. To understand these in-depth, follow the diagram below.

```python
7 + 1.96*(2)
```

```python
plt.figure(figsize=(12,8))
plt.xlim(0,25)
plt.plot(np.arange(0, 23, 0.01), norm.pdf(np.arange(0, 23, 0.01), 7, 2),'orange')
plt.fill_between(np.arange(10.92, 25, 0.01), norm.pdf(np.arange(10.92, 25, 0.01), 7, 2), 0, 
                 color='red',       
                 alpha=0.5)
plt.fill_between(np.arange(0, 10.92, 0.01), norm.pdf(np.arange(0, 10.92, 0.01), 12, 2), 0, 
                 color='orange',       
                 alpha=0.5)
plt.fill_between(np.arange(6.95, 7.05, 0.01), norm.pdf(np.arange(6.95, 7.05, 0.01), 7, 2), 0, 
                 color='green',       
                 alpha=0.3)

plt.plot(np.arange(0, 23, 0.01), norm.pdf(np.arange(0, 23, 0.01), 12, 2),'orange')
plt.fill_between(np.arange(11.95, 12.05, 0.01), norm.pdf(np.arange(11.95, 12.05, 0.01), 12, 2), 0, 
                 color='green',       
                 alpha=0.3)
```

* From the above diagram, let us consider a right tailed hypothesis test. The region shaded in red signifies alpha 0.05 (Rejection region). As the sample mean 12, falls in the rejection region, we fail to reject the null hypothesis.

* We also observe the region shaded orange, which shows type-II error region. Observe that the chance of Type-II error is high with this significance level.

```python
plt.figure(figsize=(12,8))
plt.xlim(0,25)
plt.plot(np.arange(0, 23, 0.01), norm.pdf(np.arange(0, 23, 0.01), 7, 2),'orange')
plt.fill_between(np.arange(9.5, 25, 0.01), norm.pdf(np.arange(9.5, 25, 0.01), 7, 2), 0, 
                 color='red',       
                 alpha=0.5)
plt.fill_between(np.arange(0, 9.5, 0.01), norm.pdf(np.arange(0, 9.5, 0.01), 12, 2), 0, 
                 color='orange',       
                 alpha=0.5)
plt.fill_between(np.arange(6.95, 7.05, 0.01), norm.pdf(np.arange(6.95, 7.05, 0.01), 7, 2), 0, 
                 color='green',       
                 alpha=0.3)

plt.plot(np.arange(0, 23, 0.01), norm.pdf(np.arange(0, 23, 0.01), 12, 2),'orange')
plt.fill_between(np.arange(11.95, 12.05, 0.01), norm.pdf(np.arange(11.95, 12.05, 0.01), 12, 2), 0, 
                 color='green',       
                 alpha=0.3)
```

* Now, we increase the alpha (rejection region to 0.1); increase in Type-I error rate, hence, we find a decline in chance of Type-II error.

### Hence, by conducting hypothesis tests, if we can prove that the alternate hypothesis is true, the two groups are <strong>Statistically Significant</strong>, and not attributed to any chance/ luck. If two groups are not statistically different/ significant, the claim that they are different does not hold true. 

### These form the basis of a larger group of hypothesis tests such as test of porportions, test of two sample means, ANOVA, etc. Hypothesis tests play a significant role in decision making processes, launching new products, making changes to websites, and also a huge part in the justice system.
